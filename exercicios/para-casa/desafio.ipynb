```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_olist = pd.read_csv('olist_orders_dataset (1).csv')
df_olist
```

```python
df_payment = pd.read_csv('olist_order_payments_dataset.csv')
df_payment
```

```python
df_olist_orders = pd.merge(df_olist, df_payment, on=['order_id'], how='inner')
df_olist_orders
```

```python
df_olist_orders['order_approved_at'] = pd.to_datetime(df_olist_orders['order_approved_at'])
df_olist_orders['order_delivered_carrier_date'] = pd.to_datetime(df_olist_orders['order_delivered_carrier_date'])
df_olist_orders['order_delivered_customer_date'] = pd.to_datetime(df_olist['order_delivered_customer_date'])
df_olist_orders['order_estimated_delivery_date'] = pd.to_datetime(df_olist_orders['order_estimated_delivery_date'])
```

```python
# Qual a forma de pagamento mais utilizada? 
payment_type_column = df_olist_orders['payment_type']
payment_type_counts = payment_type_column.value_counts()
most_commun = payment_type_counts.idxmax()
most_commun
```
